# Linguistic feature engineering

```python
def clean_text(document):
    """
    Removes HTML tags, URLs, mentions, hashtags, and special characters from the document.
    """
    if not isinstance(document, str):
        return ""
    # Remove HTML-like tags (e.g., class="...")
    document = re.sub(r'\b(class|src|alt)=".*?"', "", document)
    document = re.sub(r'<.*?>', "", document)  # Remove any remaining HTML tags

    # Remove URLs
    document = re.sub(r"http\S+|www\S+", "", document)

    # Remove mentions (@username)
    document = re.sub(r"@\w+", "", document)

    # Remove hashtags (#hashtag)
    document = re.sub(r"#\w+", "", document)

    # Remove non-alphanumeric characters (excluding spaces)
    document = re.sub(r"[^a-zA-Z0-9\s]", "", document)

    # Normalize whitespace
    document = re.sub(r"\s+", " ", document).strip()

    return document
```

```python
# linguistic features calculation
class Features():

  def __init__(self, social_features = True, phraseology_features= True, diversity_features = True, punct_analysis_features = True):

    self.phraseology_features = phraseology_features
    self.diversity_features = diversity_features
    self.punct_analysis_features = punct_analysis_features
    self.social_features = social_features

  def first_person(self, document):
    document = document.lower()
    tokens = word_tokenize(document)
    first_person_pronouns = {"i", "me", "my", "mine", "we", "us", "our", "ours"}
    filtered = [w for w in tokens if w.lower() in first_person_pronouns]
    return len(filtered)

  def second_person(self, document):
    document = document.lower()
    tokens = word_tokenize(document)
    second_person_pronouns = {"you", "your", "yours"}
    filtered = [w for w in tokens if w.lower() in second_person_pronouns]
    return len(filtered)

  def third_person(self, document):
    document = document.lower()
    tokens = word_tokenize(document)
    third_person_pronouns = {"he", "him", "his", "she", "her", "it", "its", "they", "them", "their", "theirs"}
    filtered = [w for w in tokens if w.lower() in third_person_pronouns]
    return len(filtered)

  def emoji_count(self,document):
    tokens = word_tokenize(document)
    emojis = [w for w in tokens if emoji.is_emoji(w)]
    return len(emojis)
  def mention_count(self, document):
    tokens = word_tokenize(document)
    filtered = [w for w in tokens if w.startswith("@")]
    return len(filtered)

  def hashtag_count(self, document):
    tokens = word_tokenize(document)
    filtered = [w for w in tokens if w.startswith("#")]
    return len(filtered)

  def url_count(self, document):
    tokens = word_tokenize(document)
    filtered = [w for w in tokens if w.startswith("http")]
    return len(filtered)

  def total_punc_count(self, document):

    punct_count = 0

    for char in document:

      if char in string.punctuation:

        punct_count +=1

    return punct_count

  def word_count(self, document):
    document = clean_text(document)

    tokens = word_tokenize(document)

    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit

    filtered = [w for w in tokens if nonPunct.match(w)]

    return len(filtered)

  def sentence_count(self, document):
    document = clean_text(document)

    tokens = sent_tokenize(document)

    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit

    filtered = [w for w in tokens if nonPunct.match(w)]

    return len(filtered)

  def paragraph_count(self, document):
    document = clean_text(document)

    tokens = document.splitlines()

    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit

    filtered = [w for w in tokens if nonPunct.match(w)]

    return len(filtered)

  def word_count_sent(self, document):
    document = clean_text(document)

    tokens = sent_tokenize(document)

    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit

    filtered = [w for w in tokens if nonPunct.match(w)]

    word_counts = [self.word_count(sent) for sent in filtered]

    if len(word_counts) ==0:

      return 0, 0

    mean = sum(word_counts) / len(word_counts)
    variance = sum([((x - mean) ** 2) for x in word_counts]) / len(word_counts)
    res = variance ** 0.5

    return mean,res

  def word_count_para(self, document):
    document = clean_text(document)

    tokens = document.splitlines()

    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit

    filtered = [w for w in tokens if nonPunct.match(w)]

    word_counts = [self.word_count(para) for para in filtered]

    if len(word_counts) ==0:

      return 0, 0

    mean = sum(word_counts) / len(word_counts)
    variance = sum([((x - mean) ** 2) for x in word_counts]) / len(word_counts)
    res = variance ** 0.5

    return mean, res

  def sent_count_para(self, document):
    document = clean_text(document)

    tokens = document.splitlines()

    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit

    filtered = [w for w in tokens if nonPunct.match(w)]

    sent_counts = [self.sentence_count(para) for para in filtered]

    if len(sent_counts) ==0:

      return 0, 0

    mean = sum(sent_counts) / len(sent_counts)
    variance = sum([((x - mean) ** 2) for x in sent_counts]) / len(sent_counts)
    res = variance ** 0.5

    return mean, res

  def readability_score(self, document):
    document = clean_text(document)
    if not document or len(document.split()) < 100:  # Minimum word count for readability calculation
        return 0, 0, 0


    try:
          r = Readability(document)
          fk = r.flesch_kincaid()
          f = r.flesch()
          ari = r.ari()
          return fk.score, f.score, ari.score
    except Exception as e:
          print(f"Error in readability calculation: {e}")
          return 0, 0, 0


  def lexical_richness(self, document):
    document = clean_text(document)
    if not document:
        return [None, None]

    # Convert to string if not already
    if not isinstance(document, str):
        document = str(document)

    words = document.strip().split()

    # Handle very short texts
    if len(words) <= 2:
        return [None, None]

    # Initialize lexical richness
    lex = LexicalRichness(document)
    ret_list = []

    # Determine appropriate window size
    try:
        if len(words) > 45:
            ret_list.append(lex.mattr(window_size=25))
        else:
            window_size = max(2, min(len(words) - 1, len(words) // 3))
            ret_list.append(lex.mattr(window_size=window_size))
    except ValueError:
        ret_list.append(None)

    try:
        ret_list.append(lex.mtld(threshold=0.72))
    except:
        ret_list.append(None)

    return ret_list

  def get_features_for_tweets(self, document):

    feature_row = []

    if self.phraseology_features:
      ## phraseology features
      feature_row.append(self.word_count(document))
      feature_row.append(self.sentence_count(document))
      feature_row.append(self.paragraph_count(document))

      # word count per sentence

      word_count_vals = self.word_count_sent(document)
      feature_row.append(word_count_vals[0])
      feature_row.append(word_count_vals[1])

      # word count per paragraph
      word_count_vals = self.word_count_para(document)
      feature_row.append(word_count_vals[0])
      feature_row.append(word_count_vals[1])

      # sentence count per paragraph
      sent_count_vals = self.sent_count_para(document)
      feature_row.append(sent_count_vals[0])
      feature_row.append(sent_count_vals[1])

    if self.diversity_features:
      # diversity features

      reareadability = self.readability_score(document)
      feature_row.append(reareadability[0])
      feature_row.append(reareadability[1])
      feature_row.append(reareadability[2])

      # word count per sentence
      richness = self.lexical_richness(document)
      feature_row.append(richness[0])
      feature_row.append(richness[1])

    if self.punct_analysis_features:
      ## punctuation features

      feature_row.append(self.total_punc_count(document))

    if self.social_features:
      ## social features

      feature_row.append(self.mention_count(document))
      feature_row.append(self.hashtag_count(document))
      feature_row.append(self.url_count(document))
      feature_row.append(self.first_person(document))
      feature_row.append(self.second_person(document))
      feature_row.append(self.third_person(document))
      feature_row.append(self.emoji_count(document))


    return feature_row
```

## TF-IDF

```python
def preprocess_text(document):
    document = clean_text(document)
    tokens = word_tokenize(document)
    processed_tokens = [
        lemmatizer.lemmatize(word.lower()) for word in tokens if word.lower() not in stop_words
    ]
    return " ".join(processed_tokens)
```

```python
dfTrain['processed_text'] = dfTrain['text'].apply(preprocess_text)
dfTest['processed_text'] = dfTest['text'].apply(preprocess_text)
dfValidation['processed_text'] = dfValidation['text'].apply(preprocess_text)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # max 1000 features

tfidf_train = tfidf_vectorizer.fit_transform(dfTrain['processed_text'])
tfidf_test = tfidf_vectorizer.transform(dfTest['processed_text'])
tfidf_validation = tfidf_vectorizer.transform(dfValidation['processed_text'])

tfidf_train_df = pd.DataFrame(tfidf_train.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_test_df = pd.DataFrame(tfidf_test.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_validation_df = pd.DataFrame(tfidf_validation.toarray(), columns=tfidf_vectorizer.get_feature_names_out())


print("TF-IDF for training data:")
print(tfidf_train_df.head())
```

## Word embedding

```python
import gensim.downloader as api
import numpy as np
```

```python
word2vec = api.load("word2vec-google-news-300")
print("Word2Vec model loaded.")
```

```python
def compute_average_word2vec(text, model, vector_size=300):
    text = clean_text(text)
    tokens = word_tokenize(text)
    valid_vectors = [model[word] for word in tokens if word in model]
    if not valid_vectors:
        return np.zeros(vector_size)
    return np.mean(valid_vectors, axis=0)
```

```python
dfTrain['word2vec_embedding'] = dfTrain['text'].apply(
    lambda text: compute_average_word2vec(text, word2vec)
)
dfTest['word2vec_embedding'] = dfTest['text'].apply(
    lambda text: compute_average_word2vec(text, word2vec)
)
dfValidation['word2vec_embedding'] = dfValidation['text'].apply(
    lambda text: compute_average_word2vec(text, word2vec)
)
```

```python
def compute_sentence_embeddings(df, column_name, model):
    df['clean_text'] = df[column_name].apply(clean_text)
    embeddings = model.encode(df['clean_text'].tolist(), show_progress_bar=True)

    # store it as list
    return [list(embedding) for embedding in embeddings]
```

## Word embedding

```python
def compute_average_word2vec(text, model, vector_size=300):
    text = clean_text(text)
    tokens = word_tokenize(text)
    valid_vectors = [model[word] for word in tokens if word in model]
    if not valid_vectors:
        return np.zeros(vector_size)
    return np.mean(valid_vectors, axis=0)
```
